```python
patterns_nomes = [{"POS":"NOUN"}, {"POS":"VERB"}]
patterns_adjetivos = [{"POS":"ADJ"}, {"POS":"VERB"}]
patterns_nome_verbo_adjetivo = [{"POS":"NOUN"}, {"POS":"VERB"}, {"POS":"ADJ"}]
patterns_nome_adjetivo = [{"POS": "NOUN"}, {"POS":"VERB"}]
patterns_misc_verbo = [{"POS": "MISC"}, {"POS": "VERB"}] 
patterns_person = [{"POS":"PER"}, {"POS":"VERB"}]
patterns_nome_proprio = [{"POS":"PROPN"}, {"POS":"VERB"}]


nome_verbos = textacy.extract.token_matches(doc, patterns = patterns_nomes)
adjetivo_verbos = textacy.extract.token_matches(doc, patterns = patterns_adjetivos)
nome_verbo_adjetivos = textacy.extract.token_matches(doc,patterns = patterns_nome_verbo_adjetivo)
misc_verbos = textacy.extract.token_matches(doc, patterns = patterns_misc_verbo)
person_verbos = textacy.extract.token_matches(doc, patterns = patterns_person)
proprio_verbos = textacy.extract.token_matches(doc, patterns = patterns_nome_proprio)


```

```python
# o que vem antes e depois dessas palavras

# Import the Matcher
from spacy.matcher import Matcher

# Initialize the Matcher with the shared vocabulary
matcher = Matcher(nlp.vocab)

# Create a pattern matching two tokens: "iPhone" and "X"
pattern = [{"TEXT": "indígena"}]

# Add the pattern to the matcher
matcher.add('PATTERN', [pattern], on_match=None)

# Use the matcher on the doc
matches = matcher(doc)
print("Matches:", [doc[start:end].text for match_id, start, end in matches])
```

```python
patternTwo = [{"LEMMA": "trans"}]
matcher.add("LoveTwo", [patternTwo])
matcher = matcher(doc, as_spans=True)
print(matcher)

```

```python
from spacy.matcher import PhraseMatcher


phrase_matcher = PhraseMatcher(nlp.vocab)
phrases = ['kaingang']
patterns = [nlp(doc) for doc in phrases]
phrase_matcher.add('Textos', None, *patterns)


for sent in doc.sents:
    for match_id, start, end in phrase_matcher(nlp(sent.text)):
        if nlp.vocab.strings[match_id] in ["Textos"]:
            print(sent.text)
            print(" ")
```
